# change weekly study hours column

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("< 5","2-10")
df.head()
```

# gender distribution

```python
#using seaborn library
plt.figure(figsize=(4,4)) #width, height
sns.countplot(data = df, x = "Gender")
plt.show()
```

```python
#we are not getting exact values of male and female
plt.figure(figsize=(5,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
#from the above chart we have analysed that the number of female in the data
#is more than the number of male
```

# Analysis of students score based upon their parent education

```python
#group by and mean value using aggregate
#Analysis of parents education on their children
gb = df.groupby("ParentEduc").agg({"MathScore" : 'mean',"ReadingScore" : 'mean',"WritingScore": 'mean'})
print(gb)
```

```python
#let analyse by heat map
plt.figure(figsize=(4,4))
sns.heatmap(gb, annot = True) #annot for values
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
```

```python
#from the above chart we have concluded that the education of the parents have good impact on their child scores
```

# Analysis of student result based on parent Martial Status

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore" : 'mean', "ReadingScore" : 'mean', "WritingScore" : 'mean'})
print(gb1)
```

```python
#Let's anlysis by heat map
plt.figure(figsize=(5,5))
sns.heatmap(gb1, annot = True)
plt.title("Relationship between Parent's Martial Status and Student's Score")
plt.show()
```

```python
# from the above chart we have concluded that there is no/neglible impact on the
#student's score due to their parent's martial status
```

# Distribution of ethenic group

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")]
print(groupA)
```

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()


l = ["group A", "group B", "group C", "group D", "group E"]
mylist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
plt.pie(mylist, labels = l, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
